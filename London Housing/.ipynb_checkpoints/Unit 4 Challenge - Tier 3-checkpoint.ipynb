```python
# The following solution seems a bit more elegant.
# It calculates a new dataframe (g) with % growth rate above month 1 for each value of sliced boroughs dataframe (dfb).
```

```python
# Calculate growth rate, g:

m1 = dfb[dfb.index == dfb.index.min()].T.iloc[:,0].tolist()
cols = dfb.columns
g = pd.DataFrame()
for col in range(0,len(cols)):
    g[cols[col]] = (dfb.iloc[:,col] - m1[col]) / m1[col] * 100
```

```python
# Simple plot of g:

fig, ax = plt.subplots() 

ax.set_title('% Price Growth in Last 20 Years') 
ax.set_xlabel('Date') 
ax.set_ylabel('% Growth') 

ax.plot(g) 
plt.show() 
```

```python
# Calculates and displays % price growth for the top 5 boroughs:

dtx = g.index.max() # Get most recent date.
ppb = dfb.T.sort_values(by=dtx,ascending=False)[dtx] # Get most recent avg price for each borough, sorted.
gpb = g.T.sort_values(by=dtx,ascending=False)[dtx] # Get most recent % growth for each borough, sorted.
top5 = gpb.iloc[:5] # Get top 5 boroughs & their % growth values.
top5list = top5.index.tolist() # Get borough names as list.
df5 = g[top5list] # Slice g for top 5 boroughs only.
```

```python
# Average percentage price growth for each borough over the last 20 years:

gpb
```

```python
# Average price for each borough in most recent month:

ppb
```

```python
# Simple visualization showing % growth for top 5 boroughs:

fig, ax = plt.subplots() 

ax.set_title('Top 5 Boroughs by % Price Growth in Last 20 Years') 
ax.set_xlabel('Date') 
ax.set_ylabel('% Growth') 

ax.plot(df5) 
ax.legend(top5list)
plt.show() 
```
