**1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).**

```python
# Define API query parameters.
queryParams = {
    #'column_index' : '0',
    'start_date' : '2017-01-01',
    'end_date' : '2017-12-31',
    'api_key' : API_KEY
}
```

```python
# Define function to concatenate an API query string, given: a dictionary of query parameter, separator string, and header string. Returns query string.
def Dict_to_API_Script(dict, separator='&', header='?'):
    query = header #start query with header
    i = 1 #set loop index
    for col in dict:
        if i == len(dict): separator = ''
        query += ( "{}={}".format(col, queryParams[col]) + separator )
        i += 1
    return query
```

```python
# Join API url.
url = '/'.join([
    "https://www.quandl.com/api/v3/datasets",
    database_code,
    dataset_code,
    'data.' + return_format,
    Dict_to_API_Script(queryParams)
])
```
